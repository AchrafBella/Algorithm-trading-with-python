# Relative Strength Index

```python
#imports necessary packages
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

plt.style.use('fivethirtyeight')
```

```python
#we should set a period
start_date = datetime.datetime(2021,1,1)
end_date = datetime.datetime(2021,4,30)

#downlaoding the dataa
data = yf.download('GOOG', start=start_date, end=end_date, group_by='ticker')
```

The RSI is calculated using average price gains and losses over a given period of time. The default time period is 14 periods with values bounded from 0 to 100.

to calculate RSI :

RSI  = $\frac{H}{H+B}$

    
- H: exponential moving average of gain
- B: exponential moving average of loss

```python
def RSI(data, windows_size=14):
    df = pd.DataFrame()
    df['Adj Close'] = data['Adj Close']
    df['delta'] = df['Adj Close'].diff(1)
    df['Up_moves'] = df['delta'].apply(lambda x:x if x>0 else 0)
    df['Down_moves'] = df['delta'].apply(lambda x:abs(x) if x<0 else 0)
    df['H'] = df['Up_moves'].rolling(windows_size).mean()
    df['B'] = df['Down_moves'].rolling(windows_size).mean()
    df['RSI'] = df['H'] / df['H']+df['B']
    return df
```

```python
pd.options.plotting.backend = "plotly"

df = RSI(data)
df.plot()
```
